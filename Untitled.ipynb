```python
#Importing Libraries
import cv2
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D, MaxPooling2D, Activation
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder
```

```python
images = []
targets=[]
normal = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\train\NORMAL'
pneumonia = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\train\PNEUMONIA'
```

```python
content = os.listdir(normal)

for image in content:
    try:
        image_path = normal+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images.append(resized_image)
        targets.append(0)
    except Exception as e:
        print("exception", e)      
```

```python
len(images)
```

```python
content = os.listdir(pneumonia)

for image in content:
    try:
        image_path = pneumonia+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images.append(resized_image)
        targets.append(1)
    except Exception as e:
        print("exception", e)
```

```python
images1 = []
targets1=[]
normal1 = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\test\NORMAL'
pneumonia1 = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\test\PNEUMONIA'
```

```python
content = os.listdir(normal1)

for image in content:
    try:
        image_path = normal1+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images1.append(resized_image)
        targets1.append(0)
    except Exception as e:
        print("exception", e) 
```

```python
content = os.listdir(pneumonia1)

for image in content:
    try:
        image_path = pneumonia1+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images1.append(resized_image)
        targets1.append(1)
    except Exception as e:
        print("exception", e)
```

```python
len(images1)
```

```python
X_train = np.array(images)/255.0
y_train = np.array(targets)

X_test = np.array(images1)/255.0
y_test = np.array(targets1)
```

```python
X_train=X_train.reshape(X_train.shape[0],64, 64, 1)
X_test=X_test.reshape(X_test.shape[0],64,64,1)
```

```python
#NEURAL NETWORK 1
from tensorflow.keras import datasets, layers, models
cnn = models.Sequential([
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu', input_shape=(64,64,1)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    layers.Conv2D(filters=256, kernel_size=(5,5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),
    
    layers.Flatten(),
    layers.Dense(100, activation='relu'),
    layers.Dropout(0.25),
    
    layers.Dense(50, activation='relu'),
    layers.Dropout(0.25),
    layers.Dense(2, activation='softmax')
])
```

```python
cnn.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cnn.fit(X_train,y_train,epochs=15)
```

```python
cnn.evaluate(X_test,y_test)
```

```python
cnn.save('best-one')
```
