```python
from tensorflow.keras.models import load_model
```

```python
cnn = load_model('best-five')
```

```python
images1 = []
targets1=[]
normal1 = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\test\NORMAL'
pneumonia1 = r'C:\Users\VENKATA PRANAV\Desktop\python\ML PROJECT\chest_xray\test\PNEUMONIA'
```

```python
import os
import cv2
content = os.listdir(normal1)

for image in content:
    try:
        image_path = normal1+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images1.append(resized_image)
        targets1.append(0)
    except Exception as e:
        print("exception", e) 
```

```python
content = os.listdir(pneumonia1)

for image in content:
    try:
        image_path = pneumonia1+'\\'+image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(image,(64,64))
        images1.append(resized_image)
        targets1.append(1)
    except Exception as e:
        print("exception", e)
```

```python
import numpy as np
X_test = np.array(images1)/255.0
y_test = np.array(targets1)
```

```python
X_test=X_test.reshape(X_test.shape[0],64,64,1)
```

```python
cnn.evaluate(X_test,y_test)
```

```python
y_pred = cnn.predict(X_test)
y_pred = [np.argmax(i) for i in y_pred]
```

```python
import seaborn as sn
import tensorflow
import matplotlib.pyplot as plt
cm = tensorflow.math.confusion_matrix(labels=y_test,predictions=y_pred)

#plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
